```python
from cv2 import imread
import matplotlib.pyplot as plt
```

```python
img = imread('./Numpy_wolf_crop.jpeg', 0) #Read image in grayscale and return in a numpy array
print('Original Image shape = ',img.shape)
print('Maximun pixel value (white) = ',img.max())
fig = plt.figure(figsize=(10,10))
plt.imshow(img, cmap='gray')
plt.colorbar(orientation='horizontal', ticks=[img.min(), img.max()])
```

```python
Recorte=img!=255
Recorte
```

```python
range(Recorte.shape[0])
```

```python
#Filas
PrimerVF=[0,0]
for i in range(Recorte.shape[0]):
  for j in range(Recorte.shape[1]):
    if Recorte[i,j]==True:
      if PrimerVF == [0,0]:
        PrimerVF=[i,j]
      SegunVF=[i,j]  
```

```python
print(PrimerVF[0])
print(SegunVF[0])
```

```python
#Columnas
PrimerValor=[0,0]
for i in range(Recorte.shape[1]):
  for j in range(Recorte.shape[0]):
    if Recorte[j,i]==True:
      if PrimerValor == [0,0]:
        PrimerValor=[j,i]
      SegunValor=[j,i]  
```

```python
print(PrimerValor[1])
print(SegunValor[1])
```

```python
img2=img[PrimerVF[0]:SegunVF[0],PrimerValor[1]:SegunValor[1]]
```

```python
fig = plt.figure(figsize=(10,10))
plt.imshow(img2, cmap="gray")
plt.colorbar(orientation='horizontal', ticks=[img.min(), img.max()])
```
