```python
import matplotlib.pyplot as plt 
import numpy as np 
from matplotlib.gridspec import GridSpec
from matplotlib import markers
```

```python


fig = plt.figure(figsize=(18,6))
gs = GridSpec(nrows=30,ncols=20)

#Gráfica 1 de barras 
Audi = [32,20,20,27]
BMW = [25,19,24,30]
Ferrari = [12,15,18,17]
labels = ['2005','2010','2015','2020']

x_pos = np.arange(1,2*len(labels),2)
wd = 0.6

ax1 = fig.add_subplot(gs[0:7,:13])
ax1.bar(x_pos,Audi,color='green',width=wd, label='Audi')
ax1.bar(x_pos+0.6,BMW,color='blue',width=wd, label='BMW')
ax1.bar(x_pos+1.2,Ferrari,color='red',width=wd, label='Ferrari')
ax1.tick_params(axis='x')
ax1.set_xlabel('Años')
ax1.set_ylabel('Carros')

ax1.set_xticks(x_pos+0.6)
ax1.set_xticklabels(labels)
plt.legend(loc='upper left')


#seno

x = np.linspace(0,10,100)
y = np.sin(x)
limx=[0,2*np.pi]
limy=[-1,0,1]

ax2 = fig.add_subplot(gs[12:30,:13])

ax2.plot(x,y,color='green', marker='*')

ax2.set_xlim([0,2*np.pi])
ax2.set_ylim([-1.2,1.2])

ax2.set_xlabel('Tiempo')
ax2.set_ylabel('Amplitud')
ax2.set_title('Seno')
ax2.legend()
ax2.set_yticks(limy)
ax2.set_xticks(limx)
ax2.grid(axis = 'y', color = 'black', linestyle = 'dashed')


#exponencial v log

x_1 = np.linspace(0, 10, 200)
y_1=[] 
y_2=[]
for i in x_1:
   res = np.exp(i)
   y_1.append(res) 
for i in x_1:
   res = np.log(i)
   y_2.append(res) 
limx=[0,2*np.pi]
limy=[-1,0,1]

ax3 = fig.add_subplot(gs[:,14:])
twin_ax3 = ax3.twinx() 
ax3.plot(x_1,y_1)
twin_ax3.plot(x_1,y_2)
ax3.set_yscale("log")

ax3.set_ylabel("e^x")
twin_ax3.set_ylabel("log(x)")
```
